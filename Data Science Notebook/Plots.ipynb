```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Scatter Plot

```python
# Method 1

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, "o-", color="Red")
plt.show()
```

```python
# Method 2

plt.scatter(x, y, marker="*", color="c")             # Shift + Tab -> for checking which attributes should be passed to any function
plt.show()
```

## Histogram Plot

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, histtype="stepfilled")      # bins -> used for analyzing the view of data by either increasing and decreasing it
plt.show()
```

## Sine - Cosine Plot

```python
x = np.linspace(0, 20, 1000)

plt.plot(x, np.sin(x), label="Sine")
plt.plot(x, np.cos(x), label="Cosine")

plt.legend()
plt.show()
```

## Legend() Attributes

```python
x = np.linspace(0, 20, 1000)

plt.plot(x, np.sin(x), label="Sine")
plt.plot(x, np.cos(x), label="Cosine")

plt.legend(loc="upper right", ncol=2, fancybox=True, shadow=True, )     # Loc -> location of legend, ncol -> legends column setting 
plt.show()
```

```python
x = np.linspace(0, 20, 1000)

option = input("Enter Function: Sine or Cosine : ")

if option == "Sine":
    plt.plot(x, np.sin(x), label="Sine")

else:
    plt.plot(x, np.cos(x), label="Cosine")

plt.legend(loc="upper right", ncol=2, fancybox=True, shadow=True, )     # Loc -> location of legend, ncol -> legends column setting 
plt.show()
```
